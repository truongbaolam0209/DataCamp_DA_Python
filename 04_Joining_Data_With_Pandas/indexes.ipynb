```python
sequels = pd.read_pickle(r'C:\000_TBL\001 DATA ANALYSIS\DATACAMP\PYTHON\ProjectFiles\04_Joining_Data_With_Pandas\Datasets\sequels.p')
financials = pd.read_pickle(r'C:\000_TBL\001 DATA ANALYSIS\DATACAMP\PYTHON\ProjectFiles\04_Joining_Data_With_Pandas\Datasets\financials.p')
```

```python
sequels = sequels.set_index('id')
financials = financials.set_index('id')

# display(sequels)
# display(financials)

# Merge sequels and financials on index id
sequels_fin = sequels.merge(financials, on='id', how='left')
display(sequels_fin)


# Self merge with suffixes as inner join with left on sequel and right on id
orig_seq = sequels_fin.merge(sequels_fin, how='inner', left_on='sequel', right_on='id', right_index=True, suffixes=('_org','_seq'))
display(orig_seq)

# Add calculation to subtract revenue_org from revenue_seq 
orig_seq['diff'] = orig_seq['revenue_seq'] - orig_seq['revenue_org']

display(orig_seq)

# Select the title_org, title_seq, and diff 
titles_diff = orig_seq[['title_org','title_seq','diff']]
display(titles_diff)


# Print the first rows of the sorted titles_diff
display(titles_diff.sort_values('diff', ascending=False).head())
```
