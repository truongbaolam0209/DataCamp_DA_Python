```python
licenses = pd.read_pickle(r'C:\000_TBL\001 DATA ANALYSIS\DATACAMP\PYTHON\ProjectFiles\04_Joining_Data_With_Pandas\Datasets\licenses.p')
biz_owners = pd.read_pickle(r'C:\000_TBL\001 DATA ANALYSIS\DATACAMP\PYTHON\ProjectFiles\04_Joining_Data_With_Pandas\Datasets\business_owners.p')
```

```python
licenses_owners = licenses.merge(biz_owners, on='account')
licenses_owners
```

```python
# Group the results by title then count the number of accounts
counted_df = licenses_owners.groupby('title').agg({'account':'count'})
counted_df
```

```python
sorted_df = counted_df.sort_values(by='account', ascending=False)
sorted_df
```
