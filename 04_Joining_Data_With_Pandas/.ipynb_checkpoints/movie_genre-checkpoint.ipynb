```python
movie_to_genres = pd.read_pickle(r'C:\000_TBL\001 DATA ANALYSIS\DATACAMP\PYTHON\ProjectFiles\04_Joining_Data_With_Pandas\Datasets\movie_to_genres.p')
movies = pd.read_pickle(r'C:\000_TBL\001 DATA ANALYSIS\DATACAMP\PYTHON\ProjectFiles\04_Joining_Data_With_Pandas\Datasets\movies.p')
```

```python
display(movie_to_genres)
display(movies)
```

```python
pop_movies = movies[movies['popularity'] > 200].sort_values(by='popularity', ascending=False)
display(pop_movies)

# Use right join to merge the movie_to_genres and pop_movies tables
genres_movies = movie_to_genres.merge(pop_movies, how='right', left_on='movie_id', right_on='id')
display(genres_movies)


# Count the number of genres
genre_count = genres_movies.groupby('genre').agg({'id':'count'})
display(genre_count)


# Plot a bar chart of the genre_count
genre_count.plot(kind='bar')
plt.show()
```
